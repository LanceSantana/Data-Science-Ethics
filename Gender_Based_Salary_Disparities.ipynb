```python
bioch = df[df['Dept'] == 1]
table = pd.pivot_table(data = bioch, values = 'Sal94', index = 'Gender', aggfunc= 'mean')
table.plot.bar()
plt.title("Average Salary by Gender in 1994 for the Biochem Department")
plt.xlabel("Gender, 0 is female, 1 is male")
plt.ylabel("Average Salary")
```

```python
bioch = df[df['Dept'] == 1]
table = pd.pivot_table(data = bioch, values = 'Sal95', index = 'Gender', aggfunc= 'mean')
table.plot.bar()
plt.title("Average Salary by Gender in 1995 for the Biochem Department")
plt.xlabel("Gender, 0 is female, 1 is male")
plt.ylabel("Average Salary")
```

```python
exp_df = df[df['Sal94'] > 100000]
table = pd.pivot_table(data = exp_df, values = 'Exper', index = 'Gender', aggfunc= 'mean')
table.plot.bar()
plt.title("Average Years of Experience Based on Gender for Individuals Making Over 100k")
plt.xlabel("Gender, 0 is female, 1 is male")
plt.ylabel("Average Experience (years)")
```

```python
table = pd.pivot_table(data = df, values = 'Sal94', index = 'Gender', aggfunc = 'mean')
table.loc[[0, 1]].plot(kind = 'bar')
plt.title("Average Salary by Gender in 1994")
plt.xlabel("Gender, 0 is female, 1 is male")
plt.ylabel("Average Salary")
```

```python
table = pd.pivot_table(data = df, values = 'Sal95', index = 'Gender', aggfunc = 'mean')
table.loc[[0, 1]].plot(kind = 'bar')
plt.title("Average Salary by Gender in 1995")
plt.xlabel("Gender, 0 is female, 1 is male")
plt.ylabel("Average Salary")
```

```python
df['Salary_Difference'] = df['Sal95'] - df['Sal94']
table = pd.pivot_table(data = df, values = 'Salary_Difference', index = 'Gender', aggfunc = 'mean')
table.loc[[0, 1]].plot(kind = 'bar')
plt.title("Average Salary Increase from 1994 to 1995 by Gender")
plt.xlabel("Gender, 0 is female, 1 is male")
plt.ylabel("Average Salary Increase")
```

```python
rank_distribution = df.groupby(["Gender", 'Rank']).size().unstack()
rank_distribution.plot(kind = 'bar')
plt.title("Rank Distribution by Gender")
plt.xlabel("Gender, 0 is female, 1 is male")
plt.ylabel("Number of employees")
```
